# **3. Data Quality Assessment**

## **3.1 Missing Value**

### **Tujuan**
Memeriksa keberadaan nilai yang hilang (*missing value*) pada setiap atribut dataset.

### **Analisi Dilakukan**
Nilai yang hilang dapat menyebabkan bias maupun *error* pada tahap analisis dan visualisasi selanjutnya. Sebagaimana
dijelaskan pada materi *Data Pre-Processing*,penanganan *missing value* merupakan salah satu tahap krusial sebelum data dapat digunakan untuk pemodelan maupun analisis, karena model atau perhitungan statistik pada umumnya tidak
dapat menangani nilai kosong secara langsung.


````python
# Memeriksa jumlah missing value pada setiap kolom
missing_value = df.isnull().sum()
persentase_missing = (missing_value / len(df)) * 100

ringkasan_missing = pd.DataFrame({
    "Jumlah Missing": missing_value,
    "Persentase (%)": persentase_missing.round(2)
})

ringkasan_missing
````

### **Interpretasi**
Hasil pemeriksaan menunjukkan bahwa **tidak terdapat nilai yang hilang (*missing value*)** pada seluruh atribut
dataset. Seluruh kolom memiliki jumlah dan persentase *missing value* sebesar 0. Bagi keandalan analisis selanjutnya, temuan ini berarti bahwa **tidak diperlukan strategi imputasi** (pengisian nilai kosong dengan mean, median, atau modus) maupun keputusan menghapus baris data, sehingga seluruh 1.000 baris data dapat digunakan secara
penuh tanpa risiko kehilangan informasi maupun bias akibat proses imputasi. Kondisi ini menunjukkan bahwa dataset
**Supermarket Sales** merupakan dataset yang sudah relatif bersih dari sisi kelengkapan data.

## **3.3 Pemeriksaan Konsistensi Data**

### **Tujuan**
Memeriksa konsistensi nilai pada atribut-atribut kategorikal, seperti `Branch`, `City`, `Customer type`, `Gender`, `Product line`, dan `Payment`, guna memastikan tidak terdapat variasi penulisan (misalnya perbedaan huruf besar/kecil atau kesalahan ketik) yang dapat memecah satu kategori menjadi beberapa kategori berbeda.

### **Alasan Dilakukan**
Salah satu permasalahan yang umum ditemukan pada data dunia nyata adalah adanya ketidakkonsistenan (*inconsistency*) dalam penulisan kategori atau kode. Oleh karena itu, fungsi `unique()` digunakan untuk menampilkan seluruh nilai unik pada setiap atribut kategorikal sehingga inkonsistensi dapat dideteksi sejak awal proses analisis.


````python
# Memeriksa nilai unik pada atribut-atribut kategorikal utama
kolom_kategorikal = ["Branch", "City", "Customer type", "Gender", "Product line", "Payment"]

for kolom in kolom_kategorikal:
    print(f"Kolom '{kolom}' memiliki {df[kolom].nunique()} kategori unik:")
    print(sorted(df[kolom].unique().tolist()))
    print("-" * 70)
````

### **Interpretasi**
Berdasarkan hasil pemeriksaan, seluruh atribut kategorikal menunjukkan nilai yang konsisten dan tidak mengandung
variasi penulisan yang tidak seharusnya (misalnya tidak ditemukan kategori seperti `"member"` dan `"Member"` yang
seharusnya merupakan kategori yang sama). Atribut `Branch` memiliki tiga kategori (A, B, C), `City` memiliki tiga
kategori kota, `Customer type` memiliki dua kategori (Member, Normal), `Gender` memiliki dua kategori (Male, Female),
`Product line` memiliki enam kategori lini produk, dan `Payment` memiliki tiga metode pembayaran. Bagi tahap
analisis selanjutnya, hal ini berarti proses *grouping* berdasarkan atribut kategorikal (misalnya penjualan per
`Product line`) akan menghasilkan kelompok data yang akurat, karena tidak ada kategori yang seharusnya sama namun
terhitung terpisah akibat perbedaan penulisan.

## **3.5 Pemeriksaan Outlier**

### **Tujuan**
Mendeteksi keberadaan nilai ekstrem (*outlier*) pada atribut numerik utama menggunakan visualisasi boxplot.

### **Alasan Dilakukan**
Boxplot merupakan salah satu metode visual yang efektif untuk mendeteksi *outlier* karena secara eksplisit menampilkan rentang interkuartil (IQR) beserta titik-titik data yang berada di luar batas atas maupun batas bawah
yang wajar. Sebagaimana dijelaskan pada materi Matplotlib, boxplot digunakan untuk mengetahui sebaran, pusat data,
rentang, dan kemungkinan *outlier* pada suatu atribut.


````python
# Membuat boxplot untuk beberapa atribut numerik utama guna mendeteksi outlier
kolom_numerik_cek_outlier = ["Unit price", "Quantity", "Total", "gross income", "Rating"]

fig, ax = plt.subplots(figsize=(10, 5))
boxplot_result = ax.boxplot(
    [df[kolom].dropna() for kolom in kolom_numerik_cek_outlier],
    tick_labels=kolom_numerik_cek_outlier,
    patch_artist=True
)

for box in boxplot_result["boxes"]:
    box.set_facecolor(PALETTE["secondary"])
    box.set_alpha(0.6)

ax.set_title("Pemeriksaan Outlier pada Atribut Numerik Utama", fontsize=13, fontweight="bold")
ax.set_ylabel("Nilai")
ax.grid(True, axis="y", alpha=PALETTE["grid_alpha"])

plt.tight_layout()
plt.show()
````

### **Interpretasi**

**Keberadaan Outlier**

Boxplot menunjukkan adanya sejumlah *outlier* pada atribut `Total` dan `gross income`, yang ditandai oleh titik data di atas batas atas (*upper whisker*). Secara statistik, nilai-nilai tersebut termasuk dalam kategori *outlier*.

**Evaluasi Outlier**

Meskipun teridentifikasi sebagai *outlier*, nilai-nilai tersebut masih tergolong wajar dalam konteks bisnis. Nilai transaksi yang tinggi dapat terjadi secara alami akibat kombinasi harga satuan yang tinggi, jumlah pembelian yang besar, atau keduanya. Dengan demikian, *outlier* yang ditemukan tidak mengindikasikan kesalahan pencatatan maupun kesalahan input data.

**Keputusan Penanganan**

Seluruh *outlier* dipertahankan dalam dataset karena merepresentasikan transaksi yang benar-benar terjadi. Menghapus data tersebut berpotensi menghilangkan informasi penting mengenai transaksi bernilai tinggi yang justru relevan dalam analisis penjualan. Selain itu, jumlah *outlier* relatif sedikit dibandingkan total 1.000 observasi sehingga tidak memberikan pengaruh yang signifikan terhadap keseluruhan hasil analisis.

Sementara itu, atribut `Unit price`, `Quantity`, dan `Rating` tidak menunjukkan *outlier* yang berarti. Hal ini menunjukkan bahwa variasi data terutama berasal dari nilai transaksi (`Total` dan `gross income`), bukan dari harga satuan, jumlah pembelian, maupun penilaian pelanggan.

## **4.3 Analisis Penjualan Bulanan**

### **Tujuan**
Menghitung dan menganalisis total nilai penjualan(`Total`) yang dikelompokkan berdasarkan bulan transaksi.

### **Alasan Dilakukan**
Analisis tren penjualan dari waktu ke waktu merupakan salah satu bentuk analisis hubungan sederhana antaratribut
(atribut waktu dan atribut nilai penjualan) yang sangat umum dilakukan dalam konteks bisnis, guna mengetahui

````python
# Mengelompokkan (grouping) data berdasarkan bulan transaksi dan menjumlahkan total penjualan
penjualan_bulanan = (
    df.groupby("Month")["Total"]
    .sum()
    .reindex(["January", "February", "March"])  # urutan bulan sesuai rentang data
)

penjualan_bulanan
````

### **Interpretasi**

Dataset mencakup transaksi penjualan selama tiga bulan, yaitu **Januari, Februari, dan Maret**. Hasil agregasi menunjukkan bahwa total penjualan mengalami perubahan dari bulan ke bulan, dengan penurunan pada satu periode yang kemudian diikuti oleh peningkatan pada periode berikutnya. Kondisi ini menunjukkan bahwa penjualan supermarket bersifat **fluktuatif**, sehingga belum membentuk tren yang konsisten, baik meningkat maupun menurun, selama periode pengamatan.

Pola tersebut akan divisualisasikan lebih lanjut menggunakan *line chart* pada Bab 5 agar perubahan penjualan antarbulan dapat diamati dengan lebih jelas.

**Business Insight:** Fluktuasi penjualan menunjukkan bahwa permintaan pelanggan dapat dipengaruhi oleh berbagai faktor, seperti periode promosi, jumlah hari operasional, maupun kondisi musiman. Oleh karena itu, pemantauan tren penjualan secara berkala dapat membantu perusahaan mengidentifikasi periode dengan penjualan yang menurun sehingga strategi promosi atau pengelolaan persediaan dapat disesuaikan secara lebih tepat.

## **4.4 Analisis Product Line**

### **Tujuan**
Menghitung total nilai penjualan (`Total`) untuk setiap kategori lini produk (`Product line`).

### **Alasan Dilakukan**
Analisis ini bertujuan untuk mengidentifikasi lini produk yang memberikan kontribusi penjualan tertinggi maupun terendah, sebagai bentuk analisis hubungan sederhana antara atribut kategorikal (`Product line`) dan atribut numerik

Perbandingan kontribusi penjualan ini akan divisualisasikan lebih lanjut menggunakan *bar chart* pada Bab 5 sehingga perbedaan performa setiap **Product Line** dapat diamati dengan lebih jelas.

**Business Insight:** Lini produk dengan kontribusi penjualan tertinggi dapat menjadi prioritas dalam strategi promosi dan pengelolaan persediaan karena memberikan kontribusi terbesar terhadap pendapatan. # **5. Visualisasi Data**

## **5.1 Persiapan Data Visualisasi**

### **Tujuan**
Menyiapkan data ringkasan (hasil agregasi pada subbab 4.3 dan 4.4) dalam bentuk yang siap divisualisasikan, termasukmenentukan nilai tertinggi pada masing-masing ringkasan data untuk kebutuhan anotasi.

### **Alasan Dilakukan**
Sebelum membangun visualisasi, data ringkasan perlu dipastikan sudah dalam format dan urutan yang benar. Selain itu,
nilai tertinggi pada masing-masing kelompok data perlu diidentifikasi terlebih dahulu agar proses pemberian anotasi
(`annotate()`) pada grafik dapat dilakukan secara terprogram (bukan menebak posisi secara manual).


````python
# Data untuk subplot kiri: tren penjualan bulanan
bulan_label = penjualan_bulanan.index.tolist()
nilai_penjualan_bulanan = penjualan_bulanan.values

# Menentukan bulan dengan penjualan tertinggi
idx_bulan_tertinggi = np.argmax(nilai_penjualan_bulanan)
bulan_tertinggi = bulan_label[idx_bulan_tertinggi]
nilai_bulan_tertinggi = nilai_penjualan_bulanan[idx_bulan_tertinggi]

# Data untuk subplot kanan: penjualan per product line
product_line_label = penjualan_per_product_line.index.tolist()
nilai_penjualan_product_line = penjualan_per_product_line.values

# Menentukan product line dengan penjualan tertinggi
idx_product_line_tertinggi = np.argmax(nilai_penjualan_product_line)
product_line_tertinggi = product_line_label[idx_product_line_tertinggi]
nilai_product_line_tertinggi = nilai_penjualan_product_line[idx_product_line_tertinggi]

print(f"Bulan dengan penjualan tertinggi        : {bulan_tertinggi} ({nilai_bulan_tertinggi:,.2f})")
print(f"Product line dengan penjualan tertinggi  : {product_line_tertinggi} ({nilai_product_line_tertinggi:,.2f})")
````

### **Interpretasi**

Proses identifikasi berhasil menentukan nilai tertinggi pada masing-masing ringkasan data, baik berdasarkan bulan transaksi maupun **Product Line**. Informasi ini akan digunakan pada tahap visualisasi untuk memberikan anotasi pada titik atau batang dengan nilai terbesar, sehingga data yang paling menonjol dapat dikenali dengan lebih mudah oleh pembaca.

Pendekatan ini dilakukan secara otomatis sehingga hasil identifikasi akan selalu menyesuaikan dengan isi dataset apabila terjadi perubahan atau pembaruan data.

## **5.2 Membuat Dua Subplot: Tren Penjualan Bulanan dan Penjualan per Product Line**

### **Tujuan**
Membuat satu figur visualisasi yang terdiri atas dua subplot, yaitu:

- **Subplot kiri**: *Line chart* yang menampilkan tren total penjualan bulanan.
- **Subplot kanan**: *Bar chart* yang menampilkan total penjualan berdasarkan *Product Line*.

Kedua subplot dilengkapi dengan judul, label sumbu, legenda, grid, *value label*, anotasi pada nilai tertinggi,
serta disimpan dalam format PNG (300 dpi) dan PDF.

### **Alasan Dilakukan**
Penggunaan `plt.subplots()` dengan lebih dari satu *axes* memungkinkan dua bentuk analisis (tren waktu dan perbandingan kategori) ditampilkan secara berdampingan dalam satu figur, sehingga memudahkan pembaca laporan untuk
membandingkan kedua insight sekaligus tanpa perlu berpindah gambar.

Elemen visual seperti judul, label, legend, grid,
dan *value label* ditambahkan agar grafik informatif dan dapat dibaca secara mandiri (*self-explanatory*) tanpa
memerlukan penjelasan tambahan.


````python
# Membuat figure dengan 2 subplot (1 baris, 2 kolom)
fig, (ax_line, ax_bar) = plt.subplots(1, 2, figsize=(14, 6))

# =========================================================
# SUBPLOT KIRI: Line Chart - Trend Total Penjualan Bulanan
# =========================================================
ax_line.plot(
    bulan_label,
    nilai_penjualan_bulanan,
    marker="o",
    markersize=8,
    linewidth=2.2,
    color=PALETTE["primary"],
    label="Total Penjualan"
)

ax_line.set_title("Trend Total Penjualan Bulanan", fontsize=13, fontweight="bold")
ax_line.set_xlabel("Bulan")
ax_line.set_ylabel("Total Penjualan")
ax_line.legend(loc="upper right")
ax_line.grid(True, alpha=PALETTE["grid_alpha"])
ax_line.margins(y=0.15)  # memberi ruang ekstra agar anotasi tidak terpotong

# Anotasi pada nilai penjualan bulanan tertinggi
ax_line.annotate(
    f"Tertinggi: {nilai_bulan_tertinggi:,.0f}",
    xy=(bulan_tertinggi, nilai_bulan_tertinggi),
    xytext=(0, 20),
    textcoords="offset points",
    ha="center",
    fontsize=9,
    fontweight="bold",
    color=PALETTE["highlight"],
    arrowprops=dict(arrowstyle="->", color=PALETTE["highlight"])
)

# =========================================================
# SUBPLOT KANAN: Bar Chart - Total Penjualan per Product Line
# =========================================================
warna_bar = [
    PALETTE["highlight"] if pl == product_line_tertinggi else PALETTE["secondary"]
    for pl in product_line_label
]

bar_container = ax_bar.bar(
    product_line_label,
    nilai_penjualan_product_line,
    color=warna_bar,
    label="Total Penjualan"
)

ax_bar.set_title("Total Penjualan Berdasarkan Product Line", fontsize=13, fontweight="bold")
ax_bar.set_xlabel("Product Line")
ax_bar.set_ylabel("Total Penjualan")
ax_bar.legend(loc="upper right")
ax_bar.grid(True, axis="y", alpha=PALETTE["grid_alpha"])
ax_bar.tick_params(axis="x", rotation=30)
ax_bar.margins(y=0.15)

# Value label pada setiap batang agar nilai dapat dibaca langsung tanpa melihat sumbu Y
ax_bar.bar_label(bar_container, fmt="%.0f", padding=3, fontsize=8)

# Anotasi tambahan pada product line dengan penjualan tertinggi
ax_bar.annotate(
    "Tertinggi",
    xy=(product_line_tertinggi, nilai_product_line_tertinggi),
    xytext=(0, 28),
    textcoords="offset points",
    ha="center",
    fontsize=9,
    fontweight="bold",
    color=PALETTE["highlight"],
    arrowprops=dict(arrowstyle="->", color=PALETTE["highlight"])
)

# Judul keseluruhan figure agar kedua subplot terhubung sebagai satu narasi visual
fig.suptitle("Ringkasan Performa Penjualan Supermarket", fontsize=15, fontweight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Menyimpan visualisasi ke dalam format PNG (300 dpi) dan PDF
fig.savefig("visualisasi_penjualan.png", dpi=300, bbox_inches="tight")
fig.savefig("visualisasi_penjualan.pdf", bbox_inches="tight")

plt.show()

print("Visualisasi berhasil disimpan sebagai 'visualisasi_penjualan.png' dan 'visualisasi_penjualan.pdf'.")
````

Lini produk dengan penjualan tertinggi menunjukkan kategori yang paling diminati pelanggan selama periode pengamatan, sedangkan lini produk dengan penjualan lebih rendah memiliki kontribusi yang relatif kecil. Variasi tersebut dapat dipengaruhi oleh berbagai faktor, seperti kebutuhan pelanggan, karakteristik produk, harga, maupun strategi promosi yang diterapkan pada masing-masing kategori.

**Implikasi bisnis**

Informasi ini dapat dimanfaatkan untuk menentukan prioritas pengelolaan persediaan dan strategi pemasaran. Lini produk dengan performa tinggi perlu dijaga ketersediaan stoknya agar mampu memenuhi permintaan pelanggan, sedangkan lini produk dengan performa lebih rendah dapat dievaluasi lebih lanjut untuk mengetahui faktor yang menyebabkan rendahnya kontribusi penjualan.

### **3. Insight Bisnis Keseluruhan**

Berdasarkan kedua visualisasi, diperoleh beberapa insight utama sebagai berikut.

1. **Penjualan bulanan masih berfluktuasi**. Hal ini menunjukkan bahwa permintaan pelanggan belum stabil sehingga perusahaan perlu melakukan pemantauan penjualan secara berkala untuk mengantisipasi periode dengan penjualan yang menurun.
2. **Kontribusi setiap Product Line berbeda**. Sebagian besar pendapatan supermarket berasal dari beberapa kategori produk tertentu, sehingga perusahaan perlu memahami karakteristik produk yang paling diminati pelanggan sebagai dasar dalam menentukan strategi penjualan.
3. **Strategi operasional dapat disesuaikan berdasarkan hasil analisis**. Lini produk dengan penjualan tinggi dapat diprioritaskan dalam pengelolaan persediaan dan promosi, sedangkan lini produk dengan kontribusi rendah dapat dievaluasi melalui strategi harga, promosi, maupun penempatan produk agar penjualannya meningkat.
4. **Visualisasi mempermudah identifikasi pola bisnis**. Penggunaan line chart membantu melihat perubahan penjualan dari waktu ke waktu, sedangkan bar chart memudahkan perbandingan kontribusi antar **Product Line**. Kombinasi kedua visualisasi memberikan gambaran yang lebih komprehensif mengenai kondisi penjualan supermarket selama periode pengamatan.

# **7. Kesimpulan**

## **A. Kesimpulan Analisis Data**

1. **Kualitas data.** Dataset **Supermarket Sales** merupakan dataset yang sudah relatif bersih, ditandai dengan
   tidak adanya *missing value*, data duplikat, maupun inkonsistensi penulisan kategori. Satu-satunya penyesuaian
   yang dilakukan adalah konversi tipe data pada atribut `Date` menjadi `datetime` untuk mendukung analisis
   berbasis waktu.
2. **Karakteristik data.** Distribusi atribut `Total` transaksi cenderung menceng ke kanan (*right-skewed*), yang
   mengindikasikan sebagian besar transaksi bernilai relatif kecil hingga menengah, dengan sejumlah transaksi
   bernilai besar yang tetap dipertahankan karena valid secara bisnis. Sementara itu, atribut `Rating` pelanggan
   cenderung terdistribusi mendekati simetris dengan tingkat kepuasan yang secara umum tergolong baik dan konsisten.
3. **Pola waktu dan kategori.** Total penjualan bulanan pada periode Januari–Maret menunjukkan pola yang fluktuatif,
   bukan tren yang secara konsisten meningkat maupun menurun, sementara terdapat kesenjangan kontribusi penjualan
   yang jelas antarlini produk (*Product Line*).

## **B. Kesimpulan Visualisasi**

1. Penggunaan *line chart* dengan *marker* dan anotasi berhasil menonjolkan titik penjualan bulanan tertinggi secara
   jelas dan terprogram (bukan berdasarkan penempatan manual).
2. Penggunaan *bar chart* dengan *value label* dan pewarnaan pembeda berhasil mengomunikasikan perbandingan performa
   antarlini produk secara langsung tanpa memerlukan pembacaan sumbu Y secara cermat.
3. Kombinasi kedua subplot dalam satu figur, disatukan dengan *suptitle*, memberikan narasi visual yang koheren
   antara analisis tren waktu dan analisis perbandingan kategori.
4. Visualisasi telah memenuhi seluruh elemen yang dipersyaratkan (judul, label sumbu, legenda, grid, ukuran figure,
   anotasi) serta berhasil diekspor ke dalam format PNG (300 dpi) dan PDF untuk keperluan pelaporan maupun publikasi.

## **C. Rekomendasi**

Berdasarkan hasil analisis dan visualisasi, berikut rekomendasi yang realistis dan dapat ditindaklanjuti oleh manajemen supermarket:

1. **Evaluasi periode penjualan rendah.** Melakukan investigasi lebih lanjut terhadap penyebab penurunan penjualan pada bulan Februari, misalnya dengan membandingkan kalender promosi, hari libur, maupun data eksternal lain (jika tersedia), guna merancang intervensi yang tepat sasaran.
2. **Prioritaskan lini produk unggulan.** Mengalokasikan sumber daya pemasaran dan memastikan ketersediaan stok yang optimal untuk lini produk dengan kontribusi penjualan tertinggi, sambil tetap memantau performa lini produk lainnya secara berkala.
3. **Pengumpulan data tambahan.** Mengingat dataset ini hanya mencakup periode tiga bulan,disarankan untuk melakukan analisis lanjutan dengan cakupan waktu yang lebih panjang (misalnya data tahunan) agar pola musiman dapat diidentifikasi dengan lebih meyakinkan, alih-alih hanya berdasarkan tiga titik data bulanan.
4. **Pertimbangkan segmentasi pelanggan.** Mengingat dataset turut mencatat `Customer type` dan `Gender`, praktikum lanjutan dapat memperdalam analisis dengan menyertakan dimensi tersebut untuk memahami preferensi pembelian antarsegmen pelanggan secara lebih spesifik.

Secara keseluruhan, praktikum ini menunjukkan bagaimana alur kerja *data science* mulai dari *data acquisition* hingga *conclusion* dapat diterapkan secara sistematis untuk mengubah data transaksi mentah menjadi insight bisnis
yang bermakna dan dapat ditindaklanjuti (*actionable*).
